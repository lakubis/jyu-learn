```python
import numpy as np
import matplotlib.pyplot as plt


freq = np.logspace(2, 5, 101)

w = 2*np.pi*freq

V = 10* np.exp(1j*np.pi/4)

R = 470
L = 25e-3
C = 200e-9

ZL = 1j*w*L
ZC = 1/(1j*w*C)

I = V/(ZL+ZC+R)

V_R = I*R
V_L = I*ZL
V_C = I*ZC


f, (ax1,ax2) = plt.subplots(2,1,sharex=True, figsize=(10/2.54, 10/2.54))

ax1.semilogx( freq, abs(V_R), label='V_R')
ax1.semilogx(freq, abs(V_L), label='V_L')
ax1.semilogx(freq, abs(V_C) ,label='V_C')
ax1.set_ylabel('Voltage (V)')
ax1.legend()
ax1.grid()

## Bode diagram for phase can be plotted using the ax2-handle

ax2.semilogx(freq, np.angle(V_R, deg = False), label='V_R')
ax2.semilogx(freq, np.angle(V_L, deg = False), label='V_L')
ax2.semilogx(freq, np.angle(V_C, deg = False), label='V_C')
ax2.set_ylabel('Phase (rad)')
ax2.set_xlabel('Frequency (Hz)')
ax2.set_yticks(ticks = [np.pi,0,-np.pi])
ax2.legend()
ax2.grid()
```
