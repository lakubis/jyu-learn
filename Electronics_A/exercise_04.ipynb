```python
import sympy as sp
from sympy import I, Symbol, simplify, Abs, diff,latex
import numpy as np
from sympy.solvers import solve
import matplotlib.pyplot as plt
```

```python
R1 = Symbol('R_1', positive = True)
L1 = Symbol('L_1', positive = True)
C1 = Symbol('C_1', positive = True)
R2 = Symbol('R_2', positive = True)
omega = Symbol('omega', positive = True)
V1 = Symbol('V_1', positive= True)
```

```python
XL = omega*L1
XC = 1/(omega*C1)
```

```python
Z1 = R1
Z2 = I*XL
Z3 = -I*XC
Z4 = R2
```

```python
1/((1/Z3) + (1/Z4))
```

```python
Ztot = Z1 + Z2 + (1/((1/Z3) + (1/Z4)))
```

```python
latex(Ztot)
```

```python
I1 = (V1/Ztot)
```

```python
latex(I1)
```

```python
V_U = I1*(1/((1/Z3) + (1/Z4)))
```

```python
latex(V_U)
```

```python
absVU = Abs(V_U)
```

```python
latex(absVU.simplify())
```

```python
domega_absVU = diff(absVU.simplify(),omega)
print(latex(domega_absVU))
```

```python
exp = domega_absVU.simplify()
print(latex(exp))
```

```python
exp_num = exp.subs(R1,47).subs(L1,100e-3).subs(C1,20e-9).subs(R2, 10e3).subs(V1,1)
print(latex(exp_num.simplify()))
```

```python
sol_omega = solve(exp_num,omega)
```

```python
import numpy as np

freq = 1e3; #frequency in Hz
omega = 2*np.pi*freq; # frequency in rad/s
Vs = 10; # amplitude of the signal source

R1 = 100
R2 = 300
L = 10e-3
C = 0.2e-6

Z_L = 1j*omega*L
Z_C = 1/(1j*omega*C)

#initializing the matrices
X = np.zeros([2,2],dtype=np.complex_)
B = np.zeros([2,1],dtype=np.complex_)

#defining the matrix elements
X[0,0] = 1/Z_C + 1/R1 + 1/R2
X[0,1] = -1/R2
B[0] = Vs/Z_C

X[1,0] = 1/R2
X[1,1] = -1/R2 - 1/Z_L
B[1] = 0

V=np.linalg.solve(X,B)

print('V1 = {:.3f} + {:.3f}j, V2 = {:.3f} + {:.3f}j'.format(V[0][0].real, V[0][0].imag, V[1][0].real, V[1][0].imag))

print('Transfer function: |V2/Vs| = {:.3f}, phase = {:.3f} degrees'.format(np.abs(V[1][0]/Vs), np.rad2deg(np.angle(V[1][0]/Vs))))
```
