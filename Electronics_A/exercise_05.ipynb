```python
import numpy as np
import matplotlib.pyplot as plt

Vs = 10
R1 = [1000, 2000, 5000]
C = 1e-9


imax = 20

f, ax = plt.subplots(2, figsize=(10/2.54, 15/2.54))
for R in R1:
    dt = R * C / imax * 5
    t = np.array([0])
    V_C = np.array([0])
    for i in range(1,imax):

        t = np.append(t, i* dt)

        #V_C = np.append(V_C, (Vs / R + C * V_C[i - 1] / dt) / (C / dt + 1 / R))
        V_C = np.append(V_C, (Vs / R + ((C/dt) - (1/R)) * V_C[i - 1] ) / (C / dt))



    ax[0].plot(1000 * t, V_C, label = f'$R_1$ = {R} $\Omega$')
    ax[0].set_xlabel('time (ms)')
    ax[0].set_ylabel('$V_C$ (V)')


    ax[1].plot(1000 * t, 1000 * (Vs - V_C) / R, label = f'$R_1$ = {R} $\Omega$')
    ax[1].set_xlabel('time (ms)')
    ax[1].set_ylabel('I (mA)')

t_max = 1000*R1[-1]*C*5 # set the maximum timespan
ax[0].set_xlim([0,t_max])
ax[0].set_ylim([0,10])
ax[0].legend()
ax[1].set_xlim([0,t_max])
ax[1].set_ylim([0,10])
ax[1].legend()
```
