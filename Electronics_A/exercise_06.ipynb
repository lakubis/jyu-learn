```python
import numpy as np
import matplotlib.pyplot as plt

R1 = 1e3
R2= 24e3

#transfer function of an op-amp
Av = lambda w, GB: GB/(1j*w+1)

#gain-bandwidth-product
GBP = 1e5


#closed loop gain for non-inverting circuit
#G_cl_ninv = (R1+R2)/R1
G_cl_inv = -(R2/R1)

#frequency range
x = np.logspace(0,6,100)

# transfer function for non-inverting op-amp circuit
#Gninv = Av(x,GBP)/(1+Av(x,GBP)/G_cl_ninv)

# transfer function for inverting op-amp circuit
Ginv = -Av(x,GBP)/(1+(-Av(x,GBP)/G_cl_inv))


f, ax = plt.subplots(2)

#ax[0].semilogx(x,20*np.log10(abs(Gninv)), label='closed-loop gain')
ax[0].semilogx(x,20*np.log10(abs(Ginv)), label='inverted closed-loop gain')

ax[0].semilogx(x,20*np.log10(abs(Av(x,GBP))), label ='op-amp open-loop gain')


ax[0].set_xlabel('frequency [rad/s]')
ax[0].set_ylabel('gain [dB]')
ax[0].grid()
ax[0].legend()


#phase = np.arctan(np.imag(Gninv)/np.real(Gninv))/(2*np.pi)*360
phase = np.arctan(np.imag(Ginv)/np.real(Ginv))*(360/(2*np.pi))
ax[1].semilogx(x,phase)
ax[1].set_xlabel('frequency [rad/s]')
ax[1].set_ylabel('phase [degree]')
ax[1].grid()


```
