```python


import numpy as np
import matplotlib.pyplot as plt

#angular frequency range
w=np.logspace(0, 7, 1000)

# Insert constants
RC = 5.6e3
RL = 10e3
RE = 1.3e3
CE = 10e-6
beta = 120
IE = 1.04e-3
hie = beta/(40*IE)

# Calculate impedance of the capacitor
ZCE = 1/(1j*w*CE)

# Calculate the parallel values
RC_par_RL = (RC*RL)/(RC + RL)
RE_par_ZCE = (RE*ZCE)/(RE + ZCE)

# Calculate gain
gain = (-beta*RC_par_RL)/(hie + (1+beta)*RE_par_ZCE)


# the plotting starts here. You do not necessarily need to
# modify anything below this point.
f, ax = plt.subplots(2, figsize=(15/2.54, 20/2.54), dpi=300)

ax[0].semilogx(w, 20*np.log10(abs(gain)))
ax[0].set_xlabel('rad/s')
ax[0].set_ylabel('dB')
ax[0].grid()


ax[1].semilogx(w, np.angle(gain, deg=True))
ax[1].set_xlabel('rad/s')
ax[1].set_ylabel('phase [$^\circ$]')
ax[1].grid()


```
