```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SiN1 = np.loadtxt('SiliconNitride1.dat',skiprows=2)
E_SiN1 = SiN1[:,0]
T_SiN1 = SiN1[:,1]

Al1 = np.loadtxt('Al1.dat',skiprows=2)
E_Al1 = Al1[:,0]
T_Al1 = Al1[:,1]
```

```python
SiN2 = np.loadtxt('SiliconNitride2.dat',skiprows=2)
E_SiN2 = SiN2[:,0]
T_SiN2 = SiN2[:,1]

Al2 = np.loadtxt('Al2.dat',skiprows=2)
E_Al2 = Al2[:,0]
T_Al2 = Al2[:,1]
```

```python
fig1, axs1 = plt.subplots(ncols=1, nrows=2, figsize = (9,11))
axs1[0].plot(E_SiN2, T_SiN2)
axs1[0].set_xlabel("Energy (eV)", fontsize = 16)
axs1[0].set_ylabel("Transmission", fontsize = 16)
axs1[0].grid(True)
axs1[0].set_xlim([E_SiN2[0],E_SiN2[-1]])
axs1[0].set_ylim([0,1])

axs1[1].plot(E_Al2, T_Al2)
axs1[1].set_xlabel("Energy (eV)", fontsize = 16)
axs1[1].set_ylabel("Transmission", fontsize = 16)
axs1[1].grid(True)
axs1[1].set_xlim([E_Al2[0],E_Al2[-1]])
axs1[1].set_ylim([0,1])
```

```python
# Convolute

C1 = T_SiN1*T_Al1
C2 = T_SiN2*T_Al2
Be8 = np.loadtxt('8Be.dat',skiprows=2)
Be13 = np.loadtxt('13Be.dat', skiprows= 2)

Fig2, ax2 = plt.subplots(figsize = (9,5))
ax2.plot(E_Al1, C1, label = "4$\mu$m Al + 4$\mu$m Si", color = 'blue')
#ax2.plot(E_Al2, C2, label = "C2", color = 'red')
ax2.plot(Be8[:,0], Be8[:,1], label = "8 $\mu$m Be", color = 'grey')
ax2.plot(Be13[:,0], Be13[:,1], label = "13 $\mu$m Be", color = 'black')
ax2.set_xlabel("Energy (eV)", fontsize = 16)
ax2.set_ylabel("Transmission", fontsize = 16)
ax2.grid(True)
ax2.set_xlim([E_SiN2[0],E_SiN2[-1]])
ax2.set_ylim([1e-3,1e1])
ax2.set_yscale('log')
ax2.legend()
```
