```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft
```

```python
data1 = pd.read_excel("sampled_signals_Ex2_Q3.xlsx", sheet_name=0, skiprows=2)
data2 = pd.read_excel("sampled_signals_Ex2_Q3.xlsx", sheet_name=1, skiprows=2)
data3 = pd.read_excel("sampled_signals_Ex2_Q3.xlsx", sheet_name=2, skiprows=2)
data1
```

```python
signal1 = data1["signal value [mV]"].to_numpy()
signal2 = data2["signal value [mV]"].to_numpy()
signal3 = data3["signal value [mV]"].to_numpy()
time1 = data1["time"].to_numpy()
time2 = data2["time"].to_numpy()
time3 = data3["time"].to_numpy()
```

```python
dt1 = time1[1] - time1[0]
dt2 = time2[1] - time2[0]
dt3 = time3[1] - time3[0]
```

```python
fig1, axs1 = plt.subplots(nrows=3, ncols=1, figsize = (9,11))
axs1[0].plot(time1, signal1)
axs1[0].set_xlabel("time (s)", fontsize = 16)
axs1[0].set_ylabel("Signal (mV)", fontsize = 16)
axs1[0].grid(True)
axs1[0].set_xlim(time1[0], time1[-1])

axs1[1].plot(time2, signal2)
axs1[1].set_xlabel("time (s)", fontsize = 16)
axs1[1].set_ylabel("Signal (mV)", fontsize = 16)
axs1[1].grid(True)
axs1[1].set_xlim(time2[0], time2[-1])

axs1[2].plot(time3, signal3)
axs1[2].set_xlabel("time (s)", fontsize = 16)
axs1[2].set_ylabel("Signal (mV)", fontsize = 16)
axs1[2].grid(True)
axs1[2].set_xlim(time2[0], time2[-1])
```

```python
fft1 = np.abs(fft(signal1))
fft2 = np.abs(fft(signal2))
fft3 = np.abs(fft(signal3))
```

```python
fig2, axs2 = plt.subplots(nrows=3, ncols=1, figsize = (9,11))
axs2[0].plot(time1, fft1)
axs2[0].set_xlabel("Frequency (Hz)", fontsize = 16)
axs2[0].set_ylabel("Signal (mV)", fontsize = 16)
axs2[0].grid(True)
axs2[0].set_xlim(time1[0], time1[-1])

axs2[1].plot(time2, fft2)
axs2[1].set_xlabel("Frequency (Hz)", fontsize = 16)
axs2[1].set_ylabel("Signal (mV)", fontsize = 16)
axs2[1].grid(True)
axs2[1].set_xlim(time2[0], time2[-1])

axs2[2].plot(time3, fft3)
axs2[2].set_xlabel("Frequency (Hz)", fontsize = 16)
axs2[2].set_ylabel("Signal (mV)", fontsize = 16)
axs2[2].grid(True)
axs2[2].set_xlim(time2[0], time2[-1])
```
