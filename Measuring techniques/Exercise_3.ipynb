```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Vr = np.array([1,4.2, 13.6, 38.9, 80.4, 150.3])*1e-6
strain = np.array([-1.5, -8.9, -29, -75, -200, -289])

G = 1.9e-6
Rg = 120
fit_Vr = np.linspace(Vr[0],Vr[-1],100)
fit_strain = -(4*fit_Vr)/(G*(1+2*fit_Vr))
```

```python
dRg = -(4*Rg*Vr)/(1+2*Vr)
print(dRg)
```

```python
Vout = 5*Vr
print(Vout*1e6)
```

```python
fig, ax = plt.subplots(figsize = (9,5))
ax.scatter(Vr, strain, label = "Measurement data",color = 'red')
ax.plot(fit_Vr, fit_strain, label = "Fitting")
ax.grid(True)
ax.set_xlabel("$V_r$", fontsize = 16)
ax.set_ylabel("$\mu\epsilon$", fontsize = 16)
ax.set_xlim((Vr[0],Vr[-1]))
ax.set_ylim((np.min([strain[-1],fit_strain[-1]]),strain[0]))
ax.legend()
```
