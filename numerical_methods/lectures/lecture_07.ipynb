```python
import numpy as np
import matplotlib.pyplot as plt

def euler(func, x_lim, x0 = 0.0, v0 =0.0, h = 1e-5):
    """ This is a function that solves a second-order differential equation

    Args:
        func (function): _description_
        x_lim (_type_): _description_
        x0 (_type_, optional): _description_
        v0 (_type_, optional): _description_
        h (_type_, optional): _description_. Defaults to 1e-5.

    Returns:
        _type_: _description_
    """

    X = np.zeros(int((x_lim[1]-x_lim[0])/h)+1)
    V = np.zeros(int((x_lim[1]-x_lim[0])/h)+1)

    X[0] = x0
    V[0] = v0
    t = 0

    for i in range(len(X)-1):
        V[i+1] = V[i] + func(t,V[i])*h
        X[i+1] = X[i] + V[i]*h
        t += h


    return X, V
```

```python
def ddy(x,v):
    return -0.1*v - x
```

```python
t_range = [0,5]
h = 1e-3
t_plot = np.linspace(t_range[0],t_range[1],int((t_range[1]-t_range[0])/h)+1)
```

```python
X,V = euler(ddy,x_lim = [0,5],x0 = 0, v0 = 1, h = h)
```

```python
def analytical(x):
    return 100*x - 5*(x**2) + 990 * (np.exp(-0.1*x)-1)
```

```python
fig, ax = plt.subplots(figsize = (9,5))
ax.plot(t_plot, X, label = 'X')
ax.plot(t_plot, V, label = 'V')
ax.plot(t_plot, analytical(t_plot), label = 'analytical x')
ax.legend()
```
