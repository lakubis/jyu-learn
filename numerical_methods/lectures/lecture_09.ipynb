```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def gen_data(x,a,b,c):
    return a*np.exp(-b*x) + c
```

```python
def fit_data(x,a,b,c):
    return a*np.exp(-b*x) + c
```

```python
xdata = np.linspace(0,5,50)
y = gen_data(xdata, 2,1,3)
```

```python
rnd = np.random.default_rng(seed=123123123)
```

```python
y_noise = 0.2 * rnd.normal(size = len(xdata))
y_with_noise = y + y_noise
```

```python
result = curve_fit(fit_data, xdata, y_with_noise)
```

```python
plt.plot(xdata, y_with_noise, 'b.', label = "Data with noise")
plt.plot(xdata, fit_data(xdata, result[0][0], result[0][1],result[0][2]), label = "fitting")
plt.plot(xdata, y, label = "Theoretical data")
plt.grid(True)
plt.legend()
```
